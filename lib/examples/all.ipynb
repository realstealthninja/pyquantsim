## Using the library

````python
from quantsim.core import Qubit
````

````python
myqubit = Qubit()

print(myqubit)
````

Now that we have a qubit lets perform an observation

````python
myqubit.observe()
````

The qubit has collapsed into a bit

Using gates

````python
from quantsim.gates import Identity, PauliX

myqubit = Qubit()
identity = Identity()
pauliX = PauliX()
````

The Identity gate should not make any changes to our qubit

````python
print(myqubit)

print(Qubit(matrix=(identity*myqubit)))
print(Qubit(matrix=(identity*myqubit)))
````

## Circuits


Multiplying qubits and gates together gets tiresome as the number of qubits and gates tend to N. For simplifying this process
we use quantum circuits to represent these connections

````python
from quantsim.circuit import Circuit, GateComponent, Observer, QubitComponent


circuit = Circuit()

qubit = QubitComponent()
paulixcomp = GateComponent(pauliX)
identity = GateComponent(Identity())
observer = Observer()



qubit.connect_ouput(paulixcomp)
paulixcomp.connect_ouput(identity)
identity.connect_ouput(observer)


circuit.add(qubit)
circuit.add(paulixcomp)
circuit.add(identity)
circuit.add(observer)


circuit.calculate()

print(observer.value)
print(observer.end_qubit)
````
